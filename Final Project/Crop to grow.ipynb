```python
df2 = df1.drop(['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall',"label"], axis=1)
```

```python
df2
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder().fit(df2)
```

```python
onehot1 = encoder.transform(df2)
```

```python
onehot2 = pd.DataFrame(onehot1,columns = encoder.columns_)
```

```python
onehot2
```

```python
import scipy.spatial
```

```python
jaccard = scipy.spatial.distance.cdist(onehot2, onehot1, metric = 'jaccard')
```

```python
jaccard
```

```python
jaccarddis = pd.DataFrame(jaccard,columns=onehot2.index.values,index=onehot2.index.values)
```

```python
jaccarddis
```

```python
from scipy.spatial import distance
```

```python
man_Dist0 = distance.cityblock(jaccarddis[0],1)
man_Dist1 = distance.cityblock(jaccarddis[1],1)
man_Dist2 = distance.cityblock(jaccarddis[2],1)
man_Dist3 = distance.cityblock(jaccarddis[3],1)
man_Dist4 = distance.cityblock(jaccarddis[4],1)
```

```python
print(man_Dist0)
print(man_Dist1)
print(man_Dist2)
print(man_Dist3)
print(man_Dist4)
```
